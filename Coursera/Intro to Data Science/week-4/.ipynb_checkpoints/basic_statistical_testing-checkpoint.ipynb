```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('datasets/grades.csv')
df.head()
```

```python
print('There are {} rows and {} columns'.format(df.shape[0], df.shape[1]))
```

```python
# OR
df.shape
```

```python
# Those who finish it early before 2016 i.e 2015 deadline
early_finishers = df[pd.to_datetime(df['assignment1_submission']) < '2016']
early_finishers.head()
```

```python
# Those who finished late after 2015 - late Submission
late_finisher = df[pd.to_datetime(df['assignment1_submission']) > '2016']
late_finisher.head()
```

```python
# or you can reverse this
late_finisher_1 = df[~df.index.isin(early_finishers.index)]
late_finisher_1.head()
```

```python
# taking mean of these two
print(early_finishers['assignment1_grade'].mean())
print(late_finisher['assignment1_grade'].mean())
```

```python
# these two look similar
# there t-test comes in. 
# Alternative Hypothesis and Null Hypothesis
from scipy.stats import ttest_ind
```

```python
# alpha value we chose is 0.05.
ttest_ind(early_finishers['assignment1_grade'], late_finisher['assignment1_grade'])
```

```python
# probability is 0.18 which is greater than our alpha value 0f 0.05. This means that we can't reject the null hypothesis.
```
