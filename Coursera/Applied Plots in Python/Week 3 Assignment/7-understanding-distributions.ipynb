```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4,figsize=(11,11))


grid = plt.GridSpec(2, 2, wspace=0.2, hspace=0.5)

fig.suptitle('Understanding Distributions Through Sampling', fontsize=14, fontweight='bold')

x1 = np.random.normal(0, 1, 10000)
x2 = np.random.gamma(1, 0.5, 10000)
x3 = np.random.exponential(1, 10000)
x4 = np.random.uniform(0,5, 10000)

ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[0, 1:])
ax3 = plt.subplot(grid[1, :1])
ax4 = plt.subplot(grid[1, 1:])

plt.subplot(2,2,1)
plt.hist(x1, density=True, bins=20, alpha=0.5, color='blue')
ax1.title.set_text( 'x1\nNormal')


plt.subplot(2,2,2)
plt.hist(x2, density=True, bins=20, alpha=0.5, color='orange')
ax2.title.set_text('x2\nGamma')


plt.subplot(2,2,3)
plt.hist(x3, density=True, bins=20, alpha=0.5, color='green')
ax3.title.set_text('x3\nExponential')


plt.subplot(2,2,4)
plt.hist(x4, density=True, bins=20, alpha=0.5, color='pink')
ax4.title.set_text('x4\nUniform')


plt.show()
```
