```python
corpus = open('comfort_honda_accord_2008.txt').read()
```

```python
docs = corpus.split('\n')
```

```python
corpus
```

```python
docs
```

```python
X = []
for doc in docs:
    # strip removes white spacing from both side of sentence
    # rstrip removes white space from right side
    # lstring from left side
    X.append(doc.strip())
```

```python
# Sentence Text
X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer()
```

```python
input_X = vec.fit_transform(X) 
```

```python
input_X
```

```python
input_X.toarray()
```
