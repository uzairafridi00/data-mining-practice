```python
corpus = open('location_bestwestern_hotel_sfo.txt').read()
```

```python
docs = corpus.split('\n')
```

```python
corpus
```

```python
docs
```

```python
X = []
for doc in docs:
    # Replace removes two white spaces between sentences
    X.append(doc.strip().replace("  ",""))
```

```python
# Sentence Text
X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer()
```

```python
input_X = vec.fit_transform(X) 
```

```python
input_X
```

```python
input_X.toarray()
```
