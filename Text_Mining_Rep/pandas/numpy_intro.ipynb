# Array Operations

```python
a = np.array([1,32,421,21])
b = np.array([32,4,8,91])

c = a-b
print(c)

d = a*b
print(d)
```

```python
faren = np.array([0,-10,-5,-15,0])

celsius = (faren - 31) * (5/9)
print(celsius)
```

```python
celsius > -20
```

```python
A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])
print(A*B)

# if we want to do matrix product, we use the "@" sign or use the dot function
print(A@B)
```

```python
A.shape
```

```python
# Let's create an array of integers
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print(array1.dtype)

# Now let's create an array of floats
array2 = np.array([[7.1, 8.2, 9.1], [10.4, 11.2, 12.3]])
print(array2.dtype)
```

```python
array3 = array1 + array2
print(array3)
print(array3.dtype)
```

```python
print(array3.sum())
```

```python
print(array3.max())
```

```python
print(array3.min())
```

```python
from PIL import Image
from IPython.display import display
# And let's just look at the image I'm talking about
im = Image.open('logo.png')
display(im)
```

```python
array = np.array(im)
print(array.shape)
array
```

```python
mask = np.full(array.shape, 255)
mask
```

```python
modified_array = array-mask
modified_array = modified_array * -1
modified_array = modified_array.astype(np.uint8)
modified_array
```

```python
display = Image.fromarray(modified_array)
```

```python
display
```

```python
reshaped=np.reshape(modified_array,(1200,1200))
print(reshaped)
display(Image.fromarray(reshaped))
```
